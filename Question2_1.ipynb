```python
y= df1['quote'].values
x= df1[[99, 100, 114, 121, 122, 125, 126, 128, 152, 189, 199, 200, 201, 202]]
```

```python
# VIF
def vif(y, x):
    r2= sm.OLS(y, x).fit().rsquared
    vif= 1. / (1. - r2)
    return vif

for i in x:
    print(i, '\t', vif(y, df1[i]))
# vif>5代表存在多重共線性，因此將排除變數99與201
```

```python
x_c=df1[[100, 114, 121, 122, 125, 126, 128, 152, 189, 199, 200, 202]]
x_ad= sm.add_constant(x_c)
est= sm.OLS(y, x_ad)
est.fit().summary() # R-squared=0.89, F=23.57
```

```python
# AIC rule
precol=[100, 114, 121, 122, 125, 126, 128, 152, 189, 199, 200, 202]

import itertools
AIC= { }
for i in range(1, len(precol)+1):
    for j in itertools.combinations(precol, i):
        pre = x_c[list(j)]
        pre_2= sm.add_constant(pre)
        est= sm.OLS(y, pre_2)
        
        AIC[j]= est.fit().aic
```

```python
from collections import Counter
c= Counter(AIC)
c.most_common( )[::-10]
#121, 122, 128, 200, 202
```
